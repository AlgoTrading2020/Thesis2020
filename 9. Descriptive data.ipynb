```python
#Figure 21

import numpy as np
import matplotlib.pyplot as plt
import datetime

def read_data():
    df = pd.read_json(r'C:\Users\Fredrik\Documents\Datasets_uppsats\0.5percent_threshold_20200101-20200311.json')
    return df
df = read_data()

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height, i.e. its count
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

times = df.date_time.dt.time

before_opening = len(times[times < datetime.time(9,0,0)])
during = len(times[(times >= datetime.time(9,0,0)) & (times <= datetime.time(17,29,0))])
after_close = len(times[times >= datetime.time(17,30,0)])


fig, ax = plt.subplots(figsize=(8,4))

rects1 = ax.bar('00.00-08.59', before_opening)
rects2 = ax.bar('09.00-17.29', during)
rects3 = ax.bar('17.30-23.59', after_close)


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)


ax.set_ylim([0,300])

plt.show()
```

```python
#Figure 22

import numpy as np
import matplotlib.pyplot as plt

def read_data():
    df = pd.read_json(r'C:\Users\Fredrik\Documents\Datasets_uppsats\0.5percent_threshold_20200101-20200311.json')
    return df
df = read_data()

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height, i.e. its count
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

weekdays = df.date_time.dt.day_name()

monday = int(weekdays[weekdays == 'Monday'].value_counts())
tuesday = int(weekdays[weekdays == 'Tuesday'].value_counts())
wednesday = int(weekdays[weekdays == 'Wednesday'].value_counts())
thursday = int(weekdays[weekdays == 'Thursday'].value_counts())
friday = int(weekdays[weekdays == 'Friday'].value_counts())
try:
    saturday = int(weekdays[weekdays == 'Saturday'].value_counts())
except:
    saturday = 0
sunday = int(weekdays[weekdays == 'Sunday'].value_counts())

fig, ax = plt.subplots(figsize=(8,4))

rects1 = ax.bar('Monday', monday)
rects2 = ax.bar('Tuesday', tuesday)
rects3 = ax.bar('Wednesday', wednesday)
rects4 = ax.bar('Thursday', thursday)
rects5 = ax.bar('Friday', friday)
rects6 = ax.bar('Saturday', saturday)
rects7 = ax.bar('Sunday', sunday)

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)
autolabel(rects6)
autolabel(rects7)

ax.set_ylim([0,160])

plt.show()
```
